```python
import PyPDF2
import requests
from sentence_transformers import SentenceTransformer, util
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
import pandas as pd

# Load NLP Model for Semantic Matching
model = SentenceTransformer('all-MiniLM-L6-v2')

def extract_text_from_pdf(pdf_path):
    text = ""
    with open(pdf_path, "rb") as file:
        reader = PyPDF2.PdfReader(file)
        for page in range(len(reader.pages)):
            text += reader.pages[page].extract_text() + "\n"
    return text.strip()

# Define Skill Categories
skill_categories = [
    "C++", "Python", "Machine Learning", "Deep Learning", "Problem Solving",
    "Data Structures", "Research Skills", "MySQL", "Engineering Mathematics",
    "Django", "Java", "Software Development", "Cloud Computing",
    "AI", "Data Analysis", "Cybersecurity", "Full-Stack Development"
]

def extract_skills(resume_text, skills_list, model, threshold=0.6):
    extracted_skills = []
    resume_embedding = model.encode(resume_text, convert_to_tensor=True)

    for skill in skills_list:
        skill_embedding = model.encode(skill, convert_to_tensor=True)
        similarity = util.pytorch_cos_sim(resume_embedding, skill_embedding).item()

        if similarity > threshold:  # Increased threshold for higher accuracy
            extracted_skills.append(skill)

    return extracted_skills

# Function for ARM-based Job Recommendation
def find_frequent_skillsets(extracted_skills):
    # Simulated dataset of resumes with extracted skills
    dataset = [
        ["Python", "Machine Learning", "Deep Learning", "AI"],
        ["Java", "Software Development", "Cloud Computing"],
        ["Django", "Full-Stack Development", "HTML", "CSS"],
        ["Data Analysis", "SQL", "MySQL"],
        ["Cybersecurity", "AI Security", "Python"],
        extracted_skills  # Add current resume skills
    ]

    te = TransactionEncoder()
    te_array = te.fit(dataset).transform(dataset)
    df = pd.DataFrame(te_array, columns=te.columns_)

    frequent_itemsets = apriori(df, min_support=0.3, use_colnames=True)
    rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.2)

    return rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']]

# Read Resume and Extract Skills
resume_text = extract_text_from_pdf("/content/CV_Sarbojit_Biswas_March_2025.pdf")
extracted_skills = extract_skills(resume_text, skill_categories, model)

# Find ARM-based Job Matches
job_recommendations = find_frequent_skillsets(extracted_skills)

print("Extracted Skills:", extracted_skills)
print("\nRecommended Job Matches Based on ARM:")
print(job_recommendations)
```
